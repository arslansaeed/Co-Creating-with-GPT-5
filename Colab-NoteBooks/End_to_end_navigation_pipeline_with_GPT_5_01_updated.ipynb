# 📌 Workflow: Image → YOLO → MiDaS → GPT-5 → gTTS

# 📷 Image (Input)  
⬇️  
# 🟦 YOLO — **Object Detection**  
*(Detect objects, bounding boxes, labels)*  
⬇️  
# 📏 MiDaS — **Depth / Distance Estimation**  
*(Calculate relative distances for detected objects)*  
⬇️  
# 🧠 GPT-5 — **Reasoning & Inference**  
*(Understand scene & generate guidance/instructions)*  
⬇️  
# 🔊 gTTS — **Voice Output**  
*(Convert guidance text into speech audio)*  


#Capture Image

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from google.colab import files
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.9):
    js = Javascript('''
        async function takePhoto(quality) {
            const div = document.createElement('div');
            const capture = document.createElement('button');
            capture.textContent = '📸 Capture Photo';
            div.appendChild(capture);

            const video = document.createElement('video');
            video.style.display = 'block';
            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            document.body.appendChild(div);
            div.appendChild(video);
            video.srcObject = stream;
            await video.play();
            google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
            await new Promise((resolve) => capture.onclick = resolve);

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            canvas.getContext('2d').drawImage(video, 0, 0);
            stream.getVideoTracks()[0].stop();
            div.remove();
            return canvas.toDataURL('image/jpeg', quality);
        }
    ''')
    display(js)
    data = eval_js(f'takePhoto({quality})')
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

def upload_photo():
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    return filename

# --- Ask user which option ---
choice = input("Type 't' to take a photo with webcam, or 'u' to upload an image: ").strip().lower()

if choice == 't':
    try:
        image_path = take_photo()
        print(f"📸 Photo captured and saved as {image_path}")
    except Exception as e:
        print("❌ Camera failed, fallback to upload.")
        image_path = upload_photo()
        print(f"✅ Uploaded file: {image_path}")
else:
    image_path = upload_photo()
    print(f"✅ Uploaded file: {image_path}")
```

##YOLO based object detection, MiDaS based depth estimation


```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load YOLOv12x for higher accuracy
yolo = YOLO("yolo12x.pt")

# Load MiDaS for depth estimation
midas = torch.hub.load("intel-isl/MiDaS", "MiDaS_small")
midas.eval()
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")
transform = midas_transforms.small_transform

# Load image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# YOLO object detection
results = yolo(img_rgb)[0]
boxes = results.boxes

# MiDaS depth estimation
input_tensor = transform(img_rgb)
with torch.no_grad():
    depth_map = midas(input_tensor).squeeze().cpu().numpy()
depth_map_resized = cv2.resize(depth_map, (img.shape[1], img.shape[0]))

# Convert relative depth to estimated distance in meters
epsilon = 1e-6
inv_depth = 1.0 / (depth_map_resized + epsilon)
scale_factor = 3.0  # Adjust based on scene
depth_meters = inv_depth * scale_factor

img_annotated = img.copy()
for box in boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    label = yolo.names[cls_id]

    # Estimate object distance
    obj_depth = depth_meters[y1:y2, x1:x2]
    est_distance = np.median(obj_depth) if obj_depth.size > 0 else 0

    # Draw annotation
    cv2.rectangle(img_annotated, (x1, y1), (x2, y2), (0, 255, 0), 2)
    text = f"{label} ({conf:.2f}) | {est_distance:.2f}m"
    cv2.putText(img_annotated, text, (x1, max(0, y1 - 10)),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 2)
plt.figure(figsize=(18, 6))

# 1. Original
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("📷 Original")
plt.axis('off')

# 2. Depth Map
plt.subplot(1, 3, 2)
plt.imshow(depth_map_resized, cmap='plasma')
plt.title("🧠 MiDaS Depth Map")
plt.axis('off')
plt.colorbar(fraction=0.046, pad=0.04)

# 3. Annotated
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img_annotated, cv2.COLOR_BGR2RGB))
plt.title("📏 YOLOv12 Detection + Distance")
plt.axis('off')

plt.tight_layout()
plt.show()

cv2.imwrite("yolo12_midas_depth.jpg", img_annotated)

```

```python
# If running in Colab/Jupyter, install gTTS once:
# !pip -q install gTTS

import base64
import mimetypes
from openai import OpenAI
from gtts import gTTS
from IPython.display import Audio, display
import os

# ⚠️ Security: prefer env var, not hardcoding
# os.environ["AIMLAPI_KEY"] = "YOUR_KEY"
client = OpenAI(
    base_url="https://api.aimlapi.com/v1",
    api_key=os.environ.get(""),
)

def image_to_data_url(path):
    mime = mimetypes.guess_type(path)[0] or "application/octet-stream"
    with open(path, "rb") as f:
        b64 = base64.b64encode(f.read()).decode("utf-8")
    return f"data:{mime};base64,{b64}"
image_path = "/content/yolo12_midas_depth.jpg"  # your local image path
img_data_url = image_to_data_url(image_path)

response = client.chat.completions.create(
    model="gpt-5-chat-latest",   # or "gpt-4o" depending on your access
    messages=[
        {
            "role": "system",
            "content": """You are an assistant for a visually impaired person.
Analyze the attached image from a first-person perspective.

Describe the scene in natural, conversational language (like speaking to the user).
Mention any immediate obstacles and their approximate location (e.g., directly in front, to the left, to the right).
If there are annotations with distances (after a '|' symbol), include those values naturally in the description but do mentioned the distance.
Finally, suggest the safest direction to move.

⚠️ Important: Do NOT use numbered lists or bullet points. Write as if you are speaking normally to the user in one clear paragraph."""
        },
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Please describe the scene and obstacles."},
                {"type": "image_url", "image_url": {"url": img_data_url}},
            ],
        },
    ],
)

assistant_text = response.choices[0].message.content
print("Assistant:", assistant_text)

# 🔊 Convert to speech with gTTS and play in the notebook
tts_path = "/content/assistant_response.mp3"
tts = gTTS(text=assistant_text, lang="en", slow=False)  # change lang="ur" for Urdu, etc.
tts.save(tts_path)

# Auto-play in notebook
display(Audio(tts_path, autoplay=True))
```
